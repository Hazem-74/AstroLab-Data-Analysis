```python
from skyfield.api import Star , load , Topos
from skyfield.data import hipparcos
import pandas as pd
```

```python
with load.open(hipparcos.URL) as f :
    df = hipparcos.load_dataframe(f)
```

```python
planets = load ('de421.bsp') # ephemeris DE421
earth = planets['earth']
location = earth + Topos(longitude_degrees=(80 , 15, 12.17 ),
latitude_degrees=(35 , 49, 26.51))
ts = load.timescale()
```

```python
RAR=[]
DECR=[]
RAB=[]
DECB=[]
for i in range(0, 13):
    t = ts.utc(2025+2*i, 1, 31, 0, 0)
    barnards_star = Star.from_dataframe(df.loc[87937])
    astrometric = location.at(t).observe(barnards_star)
    raB , decB , distanceB = astrometric.radec()
    ref = Star.from_dataframe(df.loc[87916])
    astrometric = location.at(t).observe(ref)
    raR , decR , distanceR = astrometric.radec()
    RAR.append(raR._degrees)
    DECR.append(decR._degrees)
    RAB.append(raB._degrees)
    DECB.append(decB._degrees)
RAR = pd.DataFrame(RAR)
DECR = pd.DataFrame(DECR)
RAB = pd.DataFrame(RAB)
DECB = pd.DataFrame(DECB)
RAB.to_csv('RAB.csv')
DECB.to_csv('DECB.csv')
RAR.to_csv('RAR.csv')
DECR.to_csv('DECR.csv')
```
